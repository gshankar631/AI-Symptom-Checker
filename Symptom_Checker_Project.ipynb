```python
import pandas as pd

class SymptomChecker:
    def __init__(self, csv_path=r"F:\Project\Data\symptoms_dataset.csv"):
        self.data = pd.read_csv(csv_path)

    def check_symptoms(self, user_input):
        user_input = user_input.lower()
        for _, row in self.data.iterrows():
            if all(symptom.strip() in user_input for symptom in row["Symptoms"].split(",")):
                return {
                    "Condition": row["Condition"],
                    "Doctor": row["Doctor"]
                }
        return {"Condition": "Not Found", "Doctor": "General Physician"}
```
